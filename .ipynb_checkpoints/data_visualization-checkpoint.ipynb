### Visualizing Anscombe's quartet

The file `anscombe.npy` contains Anscombe’s quartet.
For each section of the quartet, we will:
- Plot the data as a scatter plot on the box [0, 20] × [0, 13].
- Use `scipy.stats.linregress()` to calculate the slope and intercept of the least squares regression line for the data and its correlation coefficient (the first three return values).
- Plot the least squares regression line over the scatter plot on the domain $x\in[0,20]$.
- Report the mean and variance in $x$ and $y$, the slope and intercept of the regression line, and the correlation coefficient.

```python
A = np.load("anscombe.npy")

#Get domain to work on
x = np.linspace(0,20,1000)

#Get linear regression for first section
m1,b1,corr1 = linregress(A[:,0],A[:,1])[:3]
#Plot first sectin=on
ax1 = plt.subplot(221)
ax1.set_title("Section 1")
ax1.scatter(A[:,0], A[:,1])
#Plot linear regression
ax1.plot(x, m1*x + b1)
ax1.set_xlim((0,20))
ax1.set_ylim((0,13))

#Get linear regression for second section
m2,b2,corr2 = linregress(A[:,2],A[:,3])[:3]
#Plot second section
ax2 = plt.subplot(222)
ax2.set_title("Section 2")
ax2.scatter(A[:,2],A[:,3])
#Plot linear regression
ax2.plot(x, m2*x+b2)
ax2.set_xlim((0,20))
ax2.set_ylim((0,13))

#Get linear regression for third section
m3,b3,corr3 = linregress(A[:,4],A[:,5])[:3]
#Plot third section
ax3 = plt.subplot(223)
ax3.set_title("Section 3")
ax3.scatter(A[:,4],A[:,5])
#Plot linear regression
ax3.plot(x, m3*x+b3)
ax3.set_xlim((0,20))
ax3.set_ylim((0,13))

#Get linear regression for fourth section
m4,b4,corr4 = linregress(A[:,6],A[:,7])[:3]
#Plot fourth section
ax4 = plt.subplot(224)
ax4.set_title("Section 4")
ax4.scatter(A[:,6],A[:,7])
#Plot linear regression
ax4.plot(x, m4*x+b4)
ax4.set_xlim((0,20))
ax4.set_ylim((0,13))

#Space out subplots
plt.subplots_adjust(hspace=.5)

plt.show()

#Print out data for each section

print()
print("SECTION 1")
print("Slope:", m1)
print("Intercept:", b1)
print("Correlation coeffecient:", corr1)
print("x Mean:", str(np.mean(A[:,0])) + ",", "y Mean:", np.mean(A[:,1]))
print("x Variance:", str(np.var(A[:,0]))+",", "y Variance:", np.var(A[:,1]))

print()
print("SECTION 2")
print("Slope:", m2)
print("Intercept:", b2)
print("Correlation coeffecient:", corr2)
print("x Mean:", str(np.mean(A[:,2])) + ",", "y Mean:", np.mean(A[:,3]))
print("x Variance:", str(np.var(A[:,2]))+",", "y Variance:", np.var(A[:,3]))

print()
print("SECTION 3")
print("Slope:", m3)
print("Intercept:", b3)
print("Correlation coeffecient:", corr3)
print("x Mean:", str(np.mean(A[:,4])) + ",", "y Mean:", np.mean(A[:,5]))
print("x Variance:", str(np.var(A[:,4]))+",", "y Variance:", np.var(A[:,5]))

print()
print("SECTION 4")
print("Slope:", m4)
print("Intercept:", b4)
print("Correlation coeffecient:", corr4)
print("x Mean:", str(np.mean(A[:,0])) + ",", "y Mean:", np.mean(A[:,1]))
print("x Variance:", str(np.var(A[:,0]))+",", "y Variance:", np.var(A[:,1]))




```

### Visualizing Bernstein polynomials
The $n + 1$ Bernstein basis polynomials of degree $n$ are defined as follows.

$$b_{n,v}(x) = \binom{n}{v}x^v(1-x)^{n-v},\quad v=0,\ 1,\ \ldots,\ n$$

We'll plot the first $10$ Bernstein basis polynomials ($n = 0,\ 1,\ 2,\ 3$) as small multiples on the domain $[0, 1] \times [0, 1]$.

```python
#Get domain
x = np.linspace(0,1,100)
#Counter for where to plot
k = 1

#Plot each polynomial in correct subplot
for n in range(4):
    for v in range(n + 1):
        polynomial = comb(n,v)*(x**v)*(1-x)**(n-v)
        ax = plt.subplot(4,4,k)
        ax.plot(x, polynomial)
        ax.set_ylim((0,1.2))
        if (k-1)%4 == 0:
            ax.set_ylabel(r"$n ={}$".format(n))
        k += 1
    k+=3-n

#Set titles for each supblot
plt.subplot(441).set_title(r"$v={}$".format(0))
plt.subplot(446).set_title(r"$v={}$".format(1))
plt.subplot(4,4,11).set_title(r"$v={}$".format(2))
plt.subplot(4,4,16).set_title(r"$v={}$".format(3))
plt.subplots_adjust(wspace=.5,hspace=.5)

plt.show()
```

### Visualizing Major League Baseball data

The file `MLB.npy` contains measurements from over 1,000 recent Major League Baseball players, compiled by UCLA.
Each row in the array represents a different player; the columns are the player’s height (in inches), weight (in pounds), and age (in years), in that order.

We'll create several visualizations to show the correlations between height, weight, and age in the MLB data set.

```python
height, weight, age = np.load("MLB.npy").T

#Get domain and linear regression
m,b = linregress(age,height)[:2]
x = np.linspace(20,50,1000)

#Plot height against age
plt.scatter(age, height, alpha=.5,edgecolor='none',s=10)
plt.plot(x, m*x+b)
plt.axis('equal')

#Set titles and show plot
plt.title("Athlete height by age")
plt.xlabel('Athlete age (years)')
plt.ylabel('Athlete height (inches)')
plt.show()
```

```python
#Get domain and linear regression
m,b = linregress(age,weight)[:2]
x = np.linspace(20,50,1000)

#Plot weight against age
plt.scatter(age,weight,alpha=.5,edgecolor='none',s=10)
plt.plot(x, m*x+b)

#Set titles and show plot
plt.title("Athlete weight by age")
plt.xlabel('Athlete age (years)')
plt.ylabel('Athlete weight (pounds)')
plt.show()
```

```python
#Get domain and linear regression
m,b = linregress(height,weight)[:2]
x = np.linspace(67,83,1000)

#Plot weight against height
plt.scatter(height, weight, alpha=.5, edgecolor='none', s=10)
plt.plot(x,m*x+b)

#Set titles and show plot
plt.title("Athlete weight by height")
plt.xlabel("Athlete height (inches)")
plt.ylabel("Athlete weight (pounds)")
plt.show()
```

### Visualizing earthquake data

The file `earthquakes.npy` contains data from over 17,000 earthquakes between 2000 and 2010 that were at least a 5 on the Richter scale.
Each row in the array represents a different earthquake; the columns are the earthquake’s date (as a fraction of the year), magnitude (on the Richter scale), longitude, and latitude, in that order.

We'll create plots that answer the following questions:
1. How many earthquakes happened every year?
2. How often do stronger earthquakes happen compared to weaker ones?
3. Where do earthquakes happen? Where do the strongest earthquakes happen?

```python
years, magnitudes, longitude, latitude = np.load("earthquakes.npy").T
```

```python
earthquake_nums = []
year_nums = np.arange(2000,2010)

#Get numbers of earthquakes by year
for year in year_nums:
    earthquake_nums.append(len(years[(years >= year)&(years<year+1)]))

#Plot bar graph of earthquake numbers by year
plt.title("Earthquakes by year")
plt.barh(year_nums, earthquake_nums)
plt.yticks(year_nums, year_nums)
plt.xlabel('Number of earthquakes')
plt.ylabel('Year')
plt.show()
```

```python
earthquake_nums = []
mag_nums = np.arange(10)

#Get numbers of earthquakes by magnitude 
for mag in mag_nums:
    earthquake_nums.append(len(magnitudes[(magnitudes>=mag)&(magnitudes<mag+1)]))

#Plot bar graph of number of each magnitude
plt.title("Earthquakes by magnitude")
plt.barh(mag_nums, earthquake_nums)
plt.yticks(mag_nums, mag_nums)
plt.xlabel('Number of earthquakes')
plt.ylabel('Magnitude')
plt.show()
```

```python
#Divide plot into weak, medium, and strong earthquakes
weak_lat, weak_long = latitude[np.where(magnitudes<=5)], longitude[np.where(magnitudes<=5)]
med_lat, med_long = latitude[np.where((5<magnitudes)&(magnitudes<=7.5))], longitude[np.where((5<magnitudes)&(magnitudes<=7.5))]
strong_lat, strong_long = latitude[np.where(magnitudes>7.5)], longitude[np.where(magnitudes>7.5)]

#Plot each with different coloring
plt.title("Earthquakes by location and magnitude")

plt.scatter(weak_long,weak_lat, edgecolor='none', alpha = .5, s=5, c='magenta', label = 'Magnitude <= 5')
plt.scatter(med_long, med_lat, edgecolor='none', alpha = .5, s=5, c='orange', label = '5< Magnitude <= 7.5')
plt.scatter(strong_long, strong_lat, edgecolor='none', alpha = .7, s=5, c='crimson', label = 'Magnitude > 7.5')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.legend(loc = 'upper left')
plt.axis('equal')
plt.show()
```

### Visualizing the Rosenbrock function

The _Rosenbrock function_ is defined as follows. $$f(x,y)=(1−x)^2 +100(y−x^2)^2$$
The minimum value of $f$ is $0$, which occurs at the point $(1, 1)$ at the bottom of a steep, banana-shaped valley of the function.

We'll use a heat map and a contour plot to visualize the Rosenbrock function. The plot the minimizer $(1, 1)$, and use a different sequential colormap for each visualization.

```python
#Set domain for X and Y
x = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, x)

#Create rosenbrock function
Z = (1-X)**2 + 100*(Y-X**2)**2

#Plot contour map and minimizer
plt.contour(X, Y, Z, 1000, cmap="inferno")
plt.scatter((1,),(1,), marker='x', c='red')

plt.show()
```

```python
#Set domain for X and Y
x = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, x)

#Create rosenbrock function
Z = (1-X)**2 + 100*(Y-X**2)**2

#Plot contour map and minimizer
plt.pcolormesh(X, Y, Z, cmap="viridis")
plt.scatter((1,),(1,), marker='x', c='red')

plt.show()
```

### GDP by country

The file `countries.npy` contains information from 20 different countries.
Each row in the array represents a different country; the columns are the 2015 population (in millions of people), the 2015 GDP (in billions of US dollars), the average male height (in centimeters), and the average female height (in centimeters), in that order.

We'll visualize this data set with at four plots.

```python
countries = np.array(
            ["Austria", "Bolivia", "Brazil", "China", "Finland",
             "Germany", "Hungary", "India", "Japan", "North Korea",
             "Montenegro", "Norway", "Peru", "South Korea", "Sri Lanka",
             "Switzerland", "Turkey", "United Kingdom", "United States", "Vietnam"])
population, gdp, male, female = np.load("countries.npy").T

#Sort GDPs from lowest to highest
indices = np.argsort(gdp)

#Plot bar graph of GDP
plt.barh(countries[indices], gdp[indices])
plt.title("GDP by country")
plt.ylabel('Country')
plt.xlabel('GDP (in billions of US dollars)')

plt.show()
```

```python
#Sort gdp
indices = np.argsort(gdp)

#Get linear regression for avg. height v gdp
m,b = linregress(gdp[indices],male[indices])[:2]
x = np.linspace(0,17500, 17500)

#Plot avg. height vs gdp and linear regression
plt.scatter(gdp[indices], male[indices])
plt.plot(x,m*x+b, c='green')
plt.xlabel("GDP (in billions of US dollars)")
plt.ylabel("Average male height (centimeters)")

plt.plot()
```

```python
indices = np.argsort(population)

#Get linear regression for gdp v population
m,b = linregress(population[indices],gdp[indices])[:2]
x = np.linspace(0,1400,1400)

#Plot avg. height vs gdp and linear regression
plt.scatter(population[indices], gdp[indices])
plt.plot(x,m*x+b, c='green')
plt.xlabel("Average population (in millions of people")
plt.ylabel("GDP (in billions of US dollars)")

plt.plot()
```

```python
#Get subplots
fig, ax = plt.subplots(1,2)

#Plot distribution of male height
ax[0].hist(male, bins = 10)
ax[0].set_title("Distribution of male height")
ax[0].set_xticks(np.arange(160,200,10))
#Plot distribution of female height
ax[1].hist(female, bins = 10, color='pink')
ax[1].set_title("Distribution of female height")
ax[1].set_xticks(np.arange(140,180,10))

plt.show()
```
